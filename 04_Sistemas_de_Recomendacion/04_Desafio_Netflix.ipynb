### Desafio Netflix

**Netflix prize contest**

premio: 1,000,000 de dolares al equipo que disminuyer el error cuadratico medio del sistema de recomendacin de Netflix en 10%. El algoritmo de Netflix se llamba Cinematch.

El concurso fue ganado por BellKor's Pragmatic Chaos team.

Inici en 2006 con una duracin de 5 aos. 

Slo se permita enviar una solucin al da. Se inscribieron alrededor de 40,000 equipos. No fue hasta poco antes del deadline que se logr superar la mejora del 10%.

**dataset**

- 480,189 usuarios
- 17,770 pelculas
- ratings en escala de 1 a 5 estrellas

training set: 100 millones de registros
test set: 1.5 millones de registros

Matriz de datos dispersa (98.8%)
