## 01 Intro a Grafos

**Red (Grafo)**

Una red es un conjunto de **Nodos** (vértices) conectados por **Aristas** (edges). A las aristas las llamaremos relaciones

> "En los Negocios y en la vida las **relaciones** importan ¿por qué no pensar lo mismo de las relaciones de nuestros datos?"
 -- Autor Desconocido


Los grafos son útiles para almacenar y representar relaciones de elementos (observaciones) a partir de atributos (propiedades).

**Grafos vistos como una base relacional**


- Los nodos son los registros en un grafo.
- Se puede almacenar datos como propiedades del nodo. Nodos similares pueden tener distintas propiedades.
- Se pueden asociar etiquetas a los nodos, representarían nombres de tablas. Notar que las etiquetas no tienen propiedades. 
- Las relaciones siempre tienen un tipo. Es posible que las relaciones tengan propiedades.
 

Es interesante notar que una *base relacional* carece de relaciones cuando se compara con un grafo.

**Top casos de uso**

- Detección de Fraudes
- Sistemas de Recomendación
- Optimización
