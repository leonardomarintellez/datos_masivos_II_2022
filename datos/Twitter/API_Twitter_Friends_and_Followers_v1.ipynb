**Crear una tabla con los usuarios que siguen los usuarios consultados**

```python
import pandas as pd
```

Crear dataframe con los _friends_ de cada cuenta

```python
user = banxico
user_out = "Banxico"
```

```python
id_user = user._json['id']
print("El id es: ", id_user)
name_user = user._json['screen_name']
print("El nombre de usuario es: ", name_user)
account_created_at = user._json['created_at']
print("La cuenta fue creada en: ", account_created_at)
friends_user = user._json['friends_count']
print("El número de amigos es: ",friends_user)
followers_user = user._json['followers_count']
print("El número de followers es: ", followers_user)
verified_user = user._json['verified']
print("¿La cuenta es verificada?: ", verified_user)
```

Para no saturar las consultas, se consultarán a lo máximo 200 amigos de cada cuenta.

```python
n = min(friends_user,200)
n
```

Para identificar variables del usuario se usara el sufijo _u

Para identificar variables de sus amigos, se usará la variable _f

```python
id_u = []
name_u = []
friends_u = []
followers_u = []

id_f = []
name_f = []
friends_f = []
followers_f = []

# ¿La cuenta que se sigue es verificada?
verified_f = []
# Agregar el nombre para tener un descriptivo de quien es
desc_name_f = []

for user in tweepy.Cursor(api.friends, screen_name =name_user).items(n):
    id_u.append(id_user)
    id_f.append(user._json['id'])
    
    name_u.append(name_user)
    name_f.append(user._json['screen_name'])
    
    friends_u.append(friends_user)
    friends_f.append(user._json['friends_count'])
    
    followers_u.append(followers_user)
    followers_f.append(user._json['followers_count'])
    
    verified_f.append(user._json['verified'])
    desc_name_f.append(user._json['name'])
    
```

```python
datos1 = pd.DataFrame()
datos1['id_u'] = id_u 
datos1['id_f'] = id_f 
datos1['name_u'] = name_u
datos1['name_f'] = name_f 
datos1['friends_u'] = friends_u 
datos1['friends_f'] = friends_f
datos1['followers_u'] = followers_u
datos1['followers_f'] = followers_f 
datos1['verified_f'] =  verified_f
datos1['desc_name_f'] =  desc_name_f
```

Sólo por si acaso reemplazar las comas en el descriptivo, para guardar los datos como csv

```python
datos1['desc_name_f'] = datos1.desc_name_f.str.replace(',','')
```
